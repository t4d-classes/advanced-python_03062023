Step 1. Load the cities (not the swapped version) CSV file created in exercise 4. Use only the city for the index. You can modify the cities.csv file with Excel to only have the first 10 rows.

```python
# Step 1 Solution

import pandas as pd

def clean_up_city(x):
    return x.split("[")[0].strip()

cities = pd.read_csv("cities.csv")
# cities = pd.DataFrame(cities[0:10])
cities["City"] = cities["City"].apply(clean_up_city)
cities.to_csv("cities_top10.csv", index=False)

cities = pd.read_csv("cities_top10.csv", index_col=0)
cities
```

Step 10. Use a pie chart to show the 2020 estimated population of the first 10 cities relative to the other cities in the data frame.

```python
# Step 10 Solution

from matplotlib import pyplot as plt
%matplotlib inline

fig_1 = plt.figure(1, figsize=(10, 10), facecolor="white")

ax1 = fig_1.add_subplot(111)

ax1.pie(
  cities["2020 Estimate"].values[0:10],
  labels=cities.index.values[0:10],
  autopct='%1.1f%%',
)

ax1.axis('equal')


plt.show()
```
